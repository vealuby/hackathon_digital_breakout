Файл для создания датасета из малых csv файлов

```python
import pandas as pd
import numpy as np
```

```python
df_stats = pd.read_csv("small_player_starts_train.csv")
df_emotions = pd.read_csv("emotions.csv")
```

```python
df_stats.head()
```

```python
# Фильтр реакций по 21.08.2023
df_emotions = df_emotions[(df_emotions['date'] >= '2023-08-21') & (df_emotions['date'] < '2023-08-22')]
```

```python
df_emotions[df_emotions.duplicated(subset=['user_id', 'item_id'],keep=False)].sort_values(by=['user_id','item_id'])
```

```python
#pd.merge(df_stats, df_emotions,how='left', )
```

```python
df_new_emo = df_emotions.copy(deep=True)
```

```python
df_new_emo['type'].unique()
```

```python
df_new_emo
```

```python
df_new_emo['is_v_top'] = np.where(df_new_emo['type'] == 'v_top', 1, 0)
df_new_emo['pos_emo'] = np.where(df_new_emo['type'] == 'pos_emotions', df_new_emo['emoji'], '')
df_new_emo['neg_emo'] = np.where(df_new_emo['type'] == 'neg_emotions', df_new_emo['emoji'], '')
```

```python
df_new_emo[df_new_emo.duplicated(subset=['user_id','item_id'],keep=False)].sort_values(by=['user_id','item_id'])
```

```python
df_new_emo.agg({'pos_emo': 'sum', 'is_v_top':'sum', 'neg_emo': 'sum'})
```

```python
df_new_emo = df_new_emo.groupby(by=['user_id', 'item_id']).agg({'date': 'min','pos_emo': 'sum', 'is_v_top':'sum', 'neg_emo': 'sum'}).reset_index()
```

```python
df_stats_grouped = df_stats.groupby(by=['user_id','item_id']).agg({'date': 'min', 'watch_time': 'sum', 'is_autorized':'sum'})
```

```python
df_stats_grouped = df_stats_grouped.reset_index()
```

```python
df_stats_grouped = df_stats_grouped[df_stats_grouped['is_autorized'] < 15]
```

```python
df_stats_grouped.loc[df_stats_grouped.is_autorized > 0, 'is_autorized'] = 1
```

```python
df2 = pd.merge(left=df_stats_grouped,right=df_new_emo,how='left',on=['user_id','item_id'])
```

```python
df2 = df2.rename(columns={'date_x': 'date_watch', 'date_y': 'date_reaction'})
```

```python
df_meta_video = pd.read_parquet('videos.parquet')
```

```python
df2 = pd.merge(left=df2,right=df_meta_video, how='left', on='item_id')
```

```python
df2['duration'] = df2['duration']/1000
df2['ctr.CTR_10days_21_07'] = df2['ctr.CTR_10days_21_07'].fillna(0)
df2['ctr.CTR_10days_01_08'] = df2['ctr.CTR_10days_01_08'].fillna(0)
df2['ctr.CTR_10days_10_08'] = df2['ctr.CTR_10days_10_08'].fillna(0)
df2['ctr.CTR_10days_21_08'] = df2['ctr.CTR_10days_21_08'].fillna(0)
```

```python
#df2.head()
```

```python
df_live = df2[df2['duration'] == 0]
```

```python
df_live = df_live.groupby('item_id').agg({'watch_time': 'max'}).reset_index()
```

```python
df_live = df_live.rename(columns={'watch_time': 'live_duration'})
```

```python
df2 = pd.merge(left=df2, right=df_live, on='item_id', how='left')
```

```python
df2['duration'] = np.where(~df2['live_duration'].isna(), df2['live_duration'], df2['duration'])
```

```python
df2 = df2.drop(columns=['live_duration'])
```

```python
df2.to_csv('user_video_reaction_meta.csv')
```
